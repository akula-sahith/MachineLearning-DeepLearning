```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=55)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
#Make Predictions
y_preds = model.predict(X_test)
print(y_preds)
```

```python
y_prob = model.predict_proba(X_test) #probability of each class
print(y_prob)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
score = accuracy_score(y_test,y_preds)
print(score)
cm = confusion_matrix(y_test,y_preds)
print(cm)
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
print(classification_report(y_test,y_preds))
```

Hyperparamter Tuning and Cross Validation

```python
#define the paramaters
penalties = ['l1','l2', None]
solvers = ['lbfgs','saga','liblinear','newton-cg','newton-cholesky']
c_values = [100,0.01,1,0.1]
```

```python
params = dict(penalty=penalties,C=c_values,solver=solvers)
```

```python
#Now Create GridSearchCV
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5,shuffle=False)
```

```python
from sklearn.model_selection import GridSearchCV
model = LogisticRegression()
grid = GridSearchCV(estimator=model,param_grid=params,cv=skf,scoring='accuracy',n_jobs=-1)
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_preds = grid.predict(X_test)
print(y_preds)
```

```python
score = accuracy_score(y_test,y_preds)
cm = confusion_matrix(y_test,y_preds)
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
print("Classification Report : ",classification_report(y_test,y_preds))
print("Accuracy Score : ",score)
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomcv = RandomizedSearchCV(estimator=model,param_distributions=params,cv=skf,scoring='accuracy',n_jobs=-1)
randomcv.fit(X_train, y_train)
randomcv.best_score_
```

```python
ypreds = randomcv.predict(X_test)
print(ypreds)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
score = accuracy_score(y_test,ypreds)
cm = confusion_matrix(y_test,ypreds)
sns.heatmap(cm,annot=True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
print("Classification Report : ",classification_report(y_test,ypreds))
print("Accuracy Score : ",score)
```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_samples=1000,n_informative=3,n_classes=3,n_features=10,random_state=55)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='ovr')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
model.fit(X_train,y_train)
```

```python
y_preds = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
score = accuracy_score(y_test,y_preds)
cm = confusion_matrix(y_test,y_preds)
sns.heatmap(cm,annot=True)
```
